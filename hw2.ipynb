````python
# 4
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

first_filered_indices = np.where(iris_2d[: , 2] > 1.5)
print(first_filered_indices)

final_answer = []

for ele in first_filered_indices[0]:
    if(iris_2d[ele][0] < 5.0):
        final_answer.append(ele)


for ele in final_answer:
    print(ele + 1)
````
